```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "/content/drive/MyDrive/datasets/antropometria-dataset-1 (1).csv"
```

```python
data = pd.read_csv(path)
data.head
```

```python
sns.histplot(data=data[data['sexo']==1], x='cintura', kde=False)
plt.title('Hombre Cintura Histogram')
plt.xlabel('Hombre Cintura')
plt.ylabel('Count')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10,5))
axs[0].hist(data['sistol'])
axs[0].set_title('Sistol')
axs[1].hist(data['diastol'])
axs[1].set_title('Diastol')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.hist(data['peso'])
plt.title('Weight')
plt.subplot(1, 2, 2)
plt.hist(data['cintura'])
plt.title('Waist')
plt.show()
```

```python
plt.scatter(data['peso'], data['cintura'])
plt.title('Weight vs Waist')
plt.xlabel('Weight')
plt.ylabel('Waist')
plt.show()
```

```python
plt.hist(data['cintura'], bins=15)
plt.title('Peso Histogram')
plt.xlabel('Peso')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(x='diastol', y='cintura', hue='sexo', data=data)
plt.title('Sistol vs Diastol by Sex')
plt.xlabel('Sistol')
plt.ylabel('Diastol')
plt.show()
```

```python
sns.boxplot(x='sexo', y='cintura', data=data[data['sexo']==1])
plt.title('Hombre Cintura Boxplot')
plt.xlabel('Sexo')
plt.ylabel('Hombre Cintura')
plt.show()
```

```python
hombre_data = data[data['sexo'] == 1]

sns.scatterplot(x='sistol', y='diastol', data=hombre_data)
plt.title('Mujer Sistol vs Mujer Diastol')
plt.xlabel('Mujer Sistol')
plt.ylabel('Mujer Diastol')
plt.show()
```
