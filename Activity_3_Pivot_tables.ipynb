```python
adultpath = '/content/sample_data/antropometria-dataset-2.csv'
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(adultpath)
df.head()
```

```python
pd.pivot_table(df,
               index = 'desc_ent',
               values = ['peso'],
               margins = True,
               margins_name ='Peso promedio',
               columns = 'sexo'
               )
```

```python
pd.pivot_table(df,
               index = 'desc_ent',
               values = ['peso'],
               aggfunc='median',
               margins = True,
               margins_name ='Peso promedio',
               columns = 'sexo'
               )
```

```python
path = '/content/sample_data/adult.csv'
```

```python
table = pd.read_csv(path)
table.head()
```

```python
xtable = pd.pivot_table(table,
               index = ['education','marital-status'],
               values = ['age'],
               aggfunc = 'count',
               margins = True,
               columns = 'sex'
               )
xtable
```

```python
xtable.columns
```

```python
total_count = (xtable['age',    'All'].sum() / 2)
```

```python
total_count
```

```python
xtable['Percentage'] = (xtable['age',  'All'] / total_count) * 100
xtable
```

```python
xtable = pd.pivot_table(table,
               index = ['education','marital-status'],
               values = ['hours-per-week'],
               aggfunc = 'mean',
               margins = True,
               columns = 'sex'
               )
xtable
```
